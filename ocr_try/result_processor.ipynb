```python
json_file1 = './result/train_task1_ocr.json'
json_file2 = './result/train_task2_ocr.json'
output_file = './result/train_ocr_filtered.json'
import json
with open(json_file1, 'r', encoding='utf-8') as file:
    data = json.load(file)
with open(json_file2, 'r', encoding='utf-8') as file:
    data.extend(json.load(file))
```

```python
data_new = []
for item in data:
    img_list = item['image']
    res_list = item['ocr']
    boxs_new_list = []
    txts_new_list = []
    for i in range(len(img_list)):
        result = res_list[i]
        for idx in range(len(result)):
            res = result[idx]
        try:
            result = result[0]
            boxes = [line[0] for line in result]
            txts = [line[1][0] for line in result]
            scores = [line[1][1] for line in result]
            boxs_new = []
            txts_new = []
            for i in range(len(boxes)):
                box = boxes[i]
                txt = txts[i]
                score = scores[i]
                # 清洗数据
                has_chinese = (lambda x=txt:sum([1 if u'\u4e00' <= i <= u'\u9fff' else 0 for i in x])>0)
                if score > 0.9 and has_chinese() and len(txt) > 1:
                    boxs_new.append(box)
                    txts_new.append(txt)
                
        except Exception as e:
            # print(f"error: {e}")
            pass
        boxs_new_list.append(boxs_new)
        txts_new_list.append(txts_new)
    del item['ocr']
    item['ocr_txts'] = txts_new_list
    item['ocr_boxs'] = boxs_new_list
    data_new.append(item)
    # print(boxs_new)
    # print(txts_new_list)

with open(output_file, 'w', encoding='utf-8') as file:
    json.dump(data_new, file, ensure_ascii=False, indent=4)
```
