```python
json_file = './result/train_task1_ocr.json'
case_idx = 100
import json
with open(json_file, 'r', encoding='utf-8') as file:
    data = json.load(file)
item = data[case_idx]
print(item['instruction'])
print(item['output'])
from paddleocr import draw_ocr
from PIL import Image
import os 
os.environ["CUDA_VISIBLE_DEVICES"] = "0,1"
img_dir = '/home/jch/data/www25/train/images/' if 'train' in json_file else '/home/jch/data/www25/test1/images/'

img_list = item['image']
res_list = item['ocr']
for i in range(len(img_list)):
    img_file = img_list[i]
    img_path = os.path.join(img_dir, img_file)
    result = res_list[i]
    for idx in range(len(result)):
        res = result[idx]
    # draw result
    result = result[0]
    image = Image.open(img_path).convert('RGB')
    boxes = [line[0] for line in result]
    txts = [line[1][0] for line in result]
    scores = [line[1][1] for line in result]
    im_show = draw_ocr(image, boxes, txts, scores, font_path='/home/jch/projects/UserIntent/utils/simfang.ttf')
    im_show = Image.fromarray(im_show)
    im_show.show()
```
